```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sun_beach_foreigntourist.csv')
```

```python
df
```

```python
df.shape
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,6))
plt.title('Sun and beach tourism in Colombia 2018 - 2021', fontdict={'fontweight':'bold', 'fontsize': 18} )



plt.plot(df.Mes, df.año2018, 'g.-', label = '2018')
plt.plot(df.Mes, df.año2019, 'y.-', label = '2019')
plt.plot(df.Mes, df.año2020, 'r.-', label = '2020')
plt.plot(df.Mes, df.año2021, 'b.-', label = '2021')



plt.ylabel('Number of tourists', {'fontsize': 14})
plt.xticks(df.Mes[::2])


plt.legend()

plt.savefig('sunbeach_Tourist 18-21.png', dpi=100)

plt.show()
```

Regardeless 2018 was the best year of total foreign tourist that came to Colombia by airplane, 2019 was the best year for tourism in the Caribbean region of the country. The first 3 months of 2019 made the difference with other years (25% = 47388 visitors)

Pandemic Year(2020)

```python
ax = plt.axes()
ax.set(facecolor = "black")
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x=df.Mes, y=df.año2020, saturation=1)

plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('Month',fontsize=15)
plt.ylabel('Number of tourist',fontsize=15)
plt.title('Sun and Beach tourist 2020',fontsize=15,fontweight="bold")
```

2021

```python
plt.figure(figsize=(14,6))
sns.barplot(x= df.Mes.iloc[0:9], y=df.año2021)
plt.ylabel("Number of tourist")

plt.title("Sun and Beach tourism 2021")
```

The first 2 months of 2020 was showing the best numbers of Beach&Sun tourism in Colombia but the pandemic arrived. Nowadays the numbers are incresing but the country is far from the top months of the past (February 2020).
